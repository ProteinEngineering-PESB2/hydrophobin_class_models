```python
import pandas as pd
from Bio import SeqIO
```

```python
matrix_data = []

for record in SeqIO.parse("../../raw_data/all_uniprot_data.fasta", "fasta"):

    row = [
        record.id,
        record.description,
        str(record.seq)
    ]

    matrix_data.append(row)

df_process = pd.DataFrame(data=matrix_data, columns=["id", "description", "sequence"])
df_process
```

```python
array_length = []

for index in df_process.index:
    array_length.append(len(df_process['sequence'][index]))

df_process['len_seq'] = array_length
```

```python
df_process_filter = df_process[(df_process['len_seq']>=50) & (df_process['len_seq']<1022)]
df_process_filter
```

```python
df_positives = pd.read_csv("../../results/processed_data/all_target_sequences.csv")
sequences_positive = df_positives['sequence'].unique().tolist()
```

```python
is_in_positive = []

for index in df_process_filter.index:
    seq = df_process_filter['sequence'][index]

    if seq in sequences_positive:
        is_in_positive.append(True)
    else:
        is_in_positive.append(False)
    

df_process_filter['is_in_positive'] = is_in_positive

df_process_filter = df_process_filter[df_process_filter['is_in_positive']==False]
df_process_filter
```

```python
df_process_filter = df_process_filter[["id", "sequence", "description", "len_seq"]]
df_process_filter
```

```python
from sklearn.utils import shuffle

df_process_filter = shuffle(df_process_filter)
df_process_filter
```

```python
df_process_filter = df_process_filter[:len(df_positives)]
df_process_filter
```

```python
df_process_filter['target'] = "Negative"

df_full_concat = pd.concat([df_positives, df_process_filter])
df_full_concat
```

```python
df_full_concat = df_full_concat.reset_index()
```

```python
is_canon = []

for index in df_full_concat.index:
    seq = df_full_concat['sequence'][index]

    is_only_canon = True
    for residue in seq:
        if residue not in ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']:
            is_only_canon=False
            break
    is_canon.append(is_only_canon)

df_full_concat['is_canon'] = is_canon

df_full_concat = df_full_concat[df_full_concat['is_canon'] == True]
```

```python
df_full_concat['target'].value_counts()
```

```python
df_full_concat = df_full_concat[["id", "sequence", "target"]]
```

```python
df_full_concat.to_csv("../../results/processed_data/full_dataset_with_negative_examples.csv", index=False)
```
