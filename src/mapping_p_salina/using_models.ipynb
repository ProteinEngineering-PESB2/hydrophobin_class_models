```python
import pandas as pd
from joblib import load
```

```python
model_binary = load("../../generated_models/Binary_model/svc_model.joblib")
model_hfb_cps = load("../../generated_models/HFB_CPs_model/knn_model.joblib")
model_hfb_class = load("../../generated_models/HFB_model/svc_trained_model.joblib")
```

```python
df_esme = pd.read_csv("../../results/mapping_p_salina/esme.csv")
df_esme
```

```python
df_prottrans_uniref = pd.read_csv("../../results/mapping_p_salina/prottrans_t5_uniref.csv")
df_prottrans_uniref
```

```python
df_prottrans_xlu50 = pd.read_csv("../../results/mapping_p_salina/prottrans_t5_xlu50.csv")
df_prottrans_xlu50
```

```python
id_esme = df_esme["id"]
df_esme = df_esme.drop(columns=["id"])
```

```python
predictions_binary = model_binary.predict(df_esme)
predictions_binary
```

```python
predictions_binary_prob = model_binary.predict_proba(df_esme)
predictions_binary_prob_df = pd.DataFrame(data=predictions_binary_prob, columns=["Negative", "Positive"])
predictions_binary_prob_df
```

```python
predictions_binary_prob_df["id"] = id_esme.values
predictions_binary_prob_df["is_HFB-CP"] = predictions_binary
predictions_binary_prob_df
```

```python
predictions_binary_positive = predictions_binary_prob_df[predictions_binary_prob_df["is_HFB-CP"] == "Positive"]
predictions_binary_negative = predictions_binary_prob_df[predictions_binary_prob_df["is_HFB-CP"] == "Negative"]
```

```python
predictions_positive = predictions_binary_positive[predictions_binary_positive["Positive"]>=0.98]
predictions_positive
```

```python
df_merge_uniref = df_prottrans_uniref.merge(right=predictions_positive, on="id")
df_merge_id = df_merge_uniref["id"]

df_merge_uniref = df_merge_uniref.drop(columns=["id", "is_HFB-CP", "Negative", "Positive"])
df_merge_uniref
```

```python
predictions_hfb_cps_prob = model_hfb_cps.predict_proba(df_merge_uniref)
predictions_hfb_cps = model_hfb_cps.predict(df_merge_uniref)
```

```python
df_predictions_hfb_cps = pd.DataFrame(data=predictions_hfb_cps_prob, columns=["CPs", "HFB"])
df_predictions_hfb_cps["id"] = df_merge_id.values
df_predictions_hfb_cps["predictions"] = predictions_hfb_cps

df_predictions_hfb_cps_HFB = df_predictions_hfb_cps[df_predictions_hfb_cps["predictions"] == "HFB"]
df_predictions_hfb_cps_CP = df_predictions_hfb_cps[df_predictions_hfb_cps["predictions"] == "CPs"]

df_predictions_hfb_cps["predictions"].value_counts()
```

```python
df_predictions_hfb_cps_HFB = df_predictions_hfb_cps_HFB[df_predictions_hfb_cps_HFB["HFB"] == 1]
df_predictions_hfb_cps_HFB
```

```python
df_predictions_hfb_cps_CP[df_predictions_hfb_cps_CP["CPs"] == 1]
```

```python
df_merge_xlu50 = df_prottrans_xlu50.merge(right=df_predictions_hfb_cps_HFB, on="id")
df_merge_xlu50_id = df_merge_xlu50["id"]
df_merge_xlu50_to_predict = df_merge_xlu50.drop(columns=["id", "predictions", "CPs", "HFB"])
df_merge_xlu50_to_predict
```

```python
predictions_hfb_prob = model_hfb_class.predict_proba(df_merge_xlu50_to_predict.values)
predictions_hfb = model_hfb_class.predict(df_merge_xlu50_to_predict.values)
```
