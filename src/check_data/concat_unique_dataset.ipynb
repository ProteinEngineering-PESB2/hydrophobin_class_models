```python
import pandas as pd
```

```python
df_cerato = pd.read_csv("../../results/processed_data/cerato_platanin.csv")
df_hfb = pd.read_csv("../../results/processed_data/hydrophobin.csv")
df_hfb_c1 = pd.read_csv("../../results/processed_data/hydrophobin_class_I.csv")
df_hfb_c2 = pd.read_csv("../../results/processed_data/hydrophobin_class_II.csv")
```

### First revision: all cerato not must exist in hfb, hfb1, and hfb2 dataset

```python
all_hfb = pd.concat([df_hfb, df_hfb_c1, df_hfb_c2], axis=0)
unique_hfb = all_hfb['sequence'].unique()
unique_hfb = [value for value in unique_hfb]
len(unique_hfb)
```

```python
only_cerato_matrix = []

for index in df_cerato.index:
    seq = df_cerato['sequence'][index]
    
    if seq not in unique_hfb:
        row = [
            df_cerato['id'][index],
            df_cerato['sequence'][index],
            df_cerato['description'][index]
        ]
        only_cerato_matrix.append(row)

df_only_cerato = pd.DataFrame(data=only_cerato_matrix, columns=["id", "sequence", "description"])
print(f"Len all {len(df_cerato)}, len only cerato {len(df_only_cerato)}")
```

### Filter by length

```python
array_length = []

for index in df_only_cerato.index:
    array_length.append(len(df_only_cerato['sequence'][index]))

df_only_cerato['len_seq'] = array_length
df_only_cerato = df_only_cerato[(df_only_cerato['len_seq']>=50) & (df_only_cerato['len_seq']<1022)]
df_only_cerato
```

### Processing HFB 1

```python
all_hfb = pd.concat([df_hfb, df_hfb_c2], axis=0)
unique_hfb = all_hfb['sequence'].unique().tolist()
unique_hfb_c2 = df_hfb_c2['sequence'].unique().tolist()
len(unique_hfb)
```

```python
is_in_hfb = []
is_in_hfb2 = []

for index in df_hfb_c1.index:
    seq = df_hfb_c1['sequence'][index]

    if seq in unique_hfb:
        is_in_hfb.append(True)
    else:
        is_in_hfb.append(False)
    
    if seq in unique_hfb_c2:
        is_in_hfb2.append(True)
    else:
        is_in_hfb2.append(False)

df_hfb_c1['is_in_hfb'] = is_in_hfb
df_hfb_c1['is_in_hfb_2'] = is_in_hfb2

df_hfb_c1_filter = df_hfb_c1[(df_hfb_c1["is_in_hfb"]==True) & (df_hfb_c1["is_in_hfb_2"]==False)]
df_hfb_c1_filter
```

```python
array_length = []

for index in df_hfb_c1_filter.index:
    array_length.append(len(df_hfb_c1_filter['sequence'][index]))

df_hfb_c1_filter['len_seq'] = array_length
df_hfb_c1_filter = df_hfb_c1_filter[(df_hfb_c1_filter['len_seq']>=50) & (df_hfb_c1_filter['len_seq']<1022)]
df_hfb_c1_filter
```

```python
df_hfb_c1_filter = df_hfb_c1_filter[["id", "sequence", "description", "len_seq"]]
df_hfb_c1_filter
```

### Processing HFB2

```python
all_hfb = pd.concat([df_hfb, df_hfb_c1], axis=0)
unique_hfb = all_hfb['sequence'].unique().tolist()
unique_hfb_c1 = df_hfb_c1['sequence'].unique().tolist()
len(unique_hfb)
```

```python
is_in_hfb = []
is_in_hfb1 = []

for index in df_hfb_c2.index:
    seq = df_hfb_c2['sequence'][index]

    if seq in unique_hfb:
        is_in_hfb.append(True)
    else:
        is_in_hfb.append(False)
    
    if seq in unique_hfb_c1:
        is_in_hfb1.append(True)
    else:
        is_in_hfb1.append(False)

df_hfb_c2['is_in_hfb'] = is_in_hfb
df_hfb_c2['is_in_hfb_1'] = is_in_hfb1

df_hfb_c2_filter = df_hfb_c2[(df_hfb_c2["is_in_hfb"]==True) & (df_hfb_c2["is_in_hfb_1"]==False)]
df_hfb_c2_filter
```

```python
array_length = []

for index in df_hfb_c2_filter.index:
    array_length.append(len(df_hfb_c2_filter['sequence'][index]))

df_hfb_c2_filter['len_seq'] = array_length
df_hfb_c2_filter = df_hfb_c2_filter[(df_hfb_c2_filter['len_seq']>=50) & (df_hfb_c2_filter['len_seq']<1022)]
df_hfb_c2_filter
```

```python
df_hfb_c2_filter = df_hfb_c2_filter[["id", "sequence", "description", "len_seq"]]
df_hfb_c2_filter
```

### Checking HFB

```python
array_length = []

for index in df_hfb.index:
    array_length.append(len(df_hfb['sequence'][index]))

df_hfb['len_seq'] = array_length
df_hfb = df_hfb[(df_hfb['len_seq']>=50) & (df_hfb['len_seq']<1022)]
df_hfb
```

### Concat all datasets

```python
df_hfb['target'] = "HFB"
df_hfb_c2_filter['target'] = "HFB Class II"
df_hfb_c1_filter['target'] = "HFB Class I"
df_only_cerato['target'] = "CPs"

df_full_concat = pd.concat([df_hfb, df_hfb_c1_filter, df_hfb_c2_filter, df_only_cerato], axis=0)
df_full_concat
```
