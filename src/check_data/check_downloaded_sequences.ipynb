```python
import pandas as pd
from Bio import SeqIO
```

### Process cerato plataning sequences

```python
name_file = "../../raw_data/cerato_platanin.fasta"
path_save_data = "../../results/processed_data/"
```

```python
matrix_data = []

for record in SeqIO.parse(name_file, "fasta"):

    if "platanin" in record.description:
        row = [
            record.id,
            record.description,
            str(record.seq)
        ]

        matrix_data.append(row)

df_process = pd.DataFrame(data=matrix_data, columns=["id", "description", "sequence"])
df_process.to_csv("../../results/processed_data/cerato_platanin.csv", index=False)
df_process
```
