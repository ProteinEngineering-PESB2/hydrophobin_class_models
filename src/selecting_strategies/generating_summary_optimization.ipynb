```python
import pandas as pd
import os
```

### Best binary models

```python
path_data = "../../results/results_optimization/binary_filter/"
list_dir = os.listdir(path_data)

list_data = []

for element in list_dir:
    df = pd.read_csv(f"{path_data}{element}")
    df['strategy'] = element.split(".")[0]

    list_data.append(df)

df_full = pd.concat(list_data, axis=0)
df_full.sort_values(by="test_mcc", ascending=False)
```
